```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("contabilidade teste.xlsx")
df.head()
```

```python
import pandas as pd

# Separar as colunas de saída (0 a 3)
saidas = df.iloc[1:, [0, 1, 2, 3]].copy()
saidas.columns = ['DATA', 'VALOR', 'SETOR', 'OBSERVAÇÃO']
saidas['TIPO'] = 'SAÍDA'

# Separar as colunas de entrada (4 a 7)
entradas = df.iloc[1:, [4, 5, 6, 7]].copy()
entradas.columns = ['DATA', 'VALOR', 'SETOR', 'OBSERVAÇÃO']
entradas['TIPO'] = 'ENTRADA'

# Remover linhas vazias
entradas = entradas.dropna(subset=['DATA', 'VALOR', 'SETOR', 'OBSERVAÇÃO'])

# Remover linhas vazias
saidas = saidas.dropna(subset=['DATA', 'VALOR', 'SETOR', 'OBSERVAÇÃO'])

# Unir os dois DataFrames
final_df = pd.concat([saidas, entradas], ignore_index=True)


# Exibir resultado
print(final_df)
```

```python
# Lista de setores que são considerados custo
setores_custo = ['Comissão', 'Marketing', 'Assinatura']

# Criar a nova coluna direto, sem função
final_df['CUSTO/DESPESA'] = np.where(
    (final_df['TIPO'] == 'SAÍDA') & (final_df['SETOR'].str.upper().isin([s.upper() for s in setores_custo])),
    'CUSTO',
    np.where(final_df['TIPO'] == 'SAÍDA', 'DESPESA', None)
)
```

```python
from tabulate import tabulate

# Supondo que o DataFrame final é `final_df`
print(tabulate(final_df, headers='keys', tablefmt='grid', showindex=False))
```

```python
final_df.to_excel('financeiro_formatado.xlsx', index=False)
```
